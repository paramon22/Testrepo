```python
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
stock_data = yf.Ticker("GME")
prices = stock_data.history(period="1y")
prices
```

```python
def make_graph(z, xlab, ylab, title):
    plt.plot(z)
    plt.xlabel(xlab)
    plt.ylabel(ylab)
    plt.title(title)
    plt.show()

data = prices['Close']  
xlaba = 'Date'
ylaba = 'Stock Price'
ttt = 'Historical Quotes for GameStop (GME) for the Past Trailing Year'

make_graph(data,xlaba,ylaba,ttt)
```
